## Developing a Model

### Implementation: Define a Performance Metric
What is the performance meteric with which you will determine the performance of your model? Create a function that calculates this performance metric, and then returns the score. 

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    # Your code here:
    return r2_score(y_true, y_predict)
    
```

### Implementation: Shuffle and Split Data
Split the data into the testing and training datasets. Shuffle the data as well to remove any bias in selecting the traing and test. 

```python
# Your code here
from sklearn.model_selection import train_test_split

target = boston['medv']
X = boston.drop(columns = ['medv'])

X_train, X_test, Y_train, Y_test = train_test_split(X,target, test_size = 0.2, random_state = 42)
```

----

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
# Five separate RFR here with the given max depths
from sklearn.ensemble import RandomForestRegressor

#Create models:
RFR2 = RandomForestRegressor(max_depth = 2)
RFR4 = RandomForestRegressor(max_depth = 4)
RFR6 = RandomForestRegressor(max_depth = 6)
RFR8 = RandomForestRegressor(max_depth = 8)
RFR10 = RandomForestRegressor(max_depth = 10)
```

```python
#Fit
RFR2.fit(X_train, Y_train)
RFR4.fit(X_train, Y_train)
RFR6.fit(X_train, Y_train)
RFR8.fit(X_train, Y_train)
RFR10.fit(X_train, Y_train)
```

```python
#Predict:
y_train_pred2 = RFR2.predict(X_train)
y_pred2= RFR2.predict(X_test)

y_train_pred4 = RFR4.predict(X_train)
y_pred4= RFR4.predict(X_test)

y_train_pred6 = RFR6.predict(X_train)
y_pred6= RFR6.predict(X_test)

y_train_pred8 = RFR8.predict(X_train)
y_pred8= RFR8.predict(X_test)

y_train_pred10 = RFR10.predict(X_train)
y_pred10= RFR10.predict(X_test)
```

Now, plot the score for each tree on the training set and on the testing set.

```python
# Produce a plot with the score for the testing and training for the different max depths

metric2 = performance_metric(Y_test, y_pred2)
metric4 = performance_metric(Y_test, y_pred4)
metric6 = performance_metric(Y_test, y_pred6)
metric8 = performance_metric(Y_test, y_pred8)
metric10 = performance_metric(Y_test, y_pred10)

print(metric2)
print(metric4)
print(metric6)
print(metric8)
print(metric10)
```
